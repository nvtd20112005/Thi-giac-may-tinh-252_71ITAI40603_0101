```python
import cv2 as cv                               # th vin x l nh
import matplotlib.pyplot as TenHoSV            # matplotlib i tn theo yu cu
import numpy as np                             # x l mng

# c nh (tn file mu theo yu cu, khng np nh)
img = cv.imread('hinh1.jpg')                   # c nh 1
img2 = cv.imread('hinh2.jpg')                  # c nh 2

# kim tra nh c c c khng
if img is None or img2 is None:
    print("Khong tim thay file anh")
else:
    # 2. kim tra kch thc nh
    print("Shape img:", img.shape)             # chiu nh (H, W, C)

    # 3. hin th nh (sai mu do BGR)
    TenHoSV.imshow(img)
    TenHoSV.title("Anh goc (BGR)")
    TenHoSV.axis('off')
    TenHoSV.show()

    # cch 2: chuyn BGR -> RGB
    img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    TenHoSV.imshow(img_rgb)
    TenHoSV.title("Anh dung mau (RGB)")
    TenHoSV.axis('off')
    TenHoSV.show()

    # 4. phng to nh
    img_big = cv.resize(img_rgb, None, fx=1.5, fy=1.5)
    TenHoSV.imshow(img_big)
    TenHoSV.title("Anh phong to")
    TenHoSV.axis('off')
    TenHoSV.show()

    # 5. crop nh
    crop_img = img_rgb[50:200, 100:300]         # ct vng nh
    TenHoSV.imshow(crop_img)
    TenHoSV.title("Anh crop")
    TenHoSV.axis('off')
    TenHoSV.show()

    # 6. ghp nh
    img2_rgb = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
    img2_rgb = cv.resize(img2_rgb, (img_rgb.shape[1], img_rgb.shape[0]))

    # ghp ngang
    h_concat = np.hstack((img_rgb, img2_rgb))
    TenHoSV.imshow(h_concat)
    TenHoSV.title("Ghep ngang")
    TenHoSV.axis('off')
    TenHoSV.show()

    # ghp dc
    v_concat = np.vstack((img_rgb, img2_rgb))
    TenHoSV.imshow(v_concat)
    TenHoSV.title("Ghep doc")
    TenHoSV.axis('off')
    TenHoSV.show()
```
